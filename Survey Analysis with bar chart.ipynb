# Data Science Survey Analysis


Data set: https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/labs/FinalModule_Coursera/data/Topic_Survey_Assignment.csv 

The participants had three options for each topic: Very Interested, Somewhat interested, and Not interested. 
2,233 respondents completed the survey.

Use the **artist layer of Matplotlib** to plot a bar chart to visualize the percentage of the respondents' 
interest in the different data science topics surveyed.



```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/labs/FinalModule_Coursera/data/Topic_Survey_Assignment.csv')
df.shape
```

```python
df
```

```python
df.sort_values(by= 'Very interested', ascending = False, inplace = True)
df
```

```python
# '%' calculus

total = 2233  
perc = round(df*100/total,2)
perc
```

```python
# plot using Artist Layer

ax = perc.plot(kind='bar', width=0.8, figsize=(20,8), color=['#5cb85c', '#5bc0de', '#d9534f'], fontsize=14)

ax.set_title("% Interested in Data Science by Area", fontsize = 16)
ax.set_ylabel("Percent", fontsize=14)

ax.legend(fontsize = 14)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False) 

for p in ax.patches:
    ax.annotate("{:.2%}".format(p.get_height()/100),
                xy=(p.get_x()+0.05, p.get_height()+0.15))

```
